```python
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
# Need to load in images
'''
We don't want to augument the dataset like we would in image classification.
We want all the images to be like what we want created. So upside down or sheared
would not help.
# By default ImageDataGenerator doesn't do any augumenation.
# The rescaling help to make sure the color is correct when we print the images.
'''

data_generator  = ImageDataGenerator(rescale=(1/255))

datagenerator = data_generator.flow_from_directory(r'C:\Users\awalker8\Documents\COVID_Furlow\Module 3 - GANs\data\goal',
                                                  target_size = (150,150),
                                                  batch_size = 64,
                                                  class_mode = 'binary')
```
